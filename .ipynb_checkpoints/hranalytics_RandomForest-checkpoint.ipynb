```python
# Filling Null values
#train_data['education'] = train_data['education'].cat.add_categories('No Degree')
train_data['education'].fillna('No Degree',inplace = True)
#test_data['education'] = test_data['education'].cat.add_categories('No Degree')
test_data['education'].fillna('No Degree',inplace = True)
```

```python
#LabelEncoding
category_encode = ['previous_year_rating', 'department', 'region', 'education', 'gender', 'recruitment_channel']
le = LabelEncoder()
for i in category_encode:
    train_data[i] = le.fit_transform(train_data[i])
    test_data[i] = le.transform(test_data[i])
```

```python
#droping employee_id
train_data.drop('employee_id', axis=1, inplace=True)
Emp_ID = test_data['employee_id']
test_data.drop('employee_id', axis=1, inplace=True)
```

```python
# train test split
X = train_data.drop("is_promoted",axis=1)
Y = train_data["is_promoted"]
#X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=42, stratify = Y)
```

```python
def RF_tune(X_tr,y_tr):
    
    max_depth = [1, 5, 10, 50, 100, 500, 1000]
    n_estimators = [5, 10, 100, 500, 1000]
        
    # for each alpha compute cv accuracy using grid search 
    # Since we dont have sepeate CV data, using 5 fold crossvalidation on train data
    
    # Classifier to be used for grid search
    clf = RandomForestClassifier()
    
    
    # hyper parameters to be used for grid search
    parameters = dict(max_depth = max_depth, n_estimators = n_estimators)
    
    grid_clf = GridSearchCV(clf,parameters, scoring = 'f1', cv=5, n_jobs=-1,return_train_score = True)
    
    grid_clf.fit(X_tr,y_tr)

    
    tr_f1 = grid_clf.cv_results_['mean_train_score']
    cv_f1 = grid_clf.cv_results_['mean_test_score']
    depth = grid_clf.cv_results_['param_max_depth']
    estimator = grid_clf.cv_results_['param_n_estimators']
    
    opt_depth = grid_clf.best_params_['max_depth']
    opt_est = grid_clf.best_params_['n_estimators']
    
    return opt_depth,opt_est,tr_f1,cv_f1,depth,estimator
```

```python
"""def plot_hyperparam_vs_f1score(param,cv_accuracy,train_accuracy):
    plt.title('KNN Varying hyper params')
    plt.plot(param, cv_accuracy, label='Validation AUC')
    plt.plot(param, train_accuracy, label='Train AUC')
    plt.legend()
    plt.xlabel('Hyper Parameter')
    plt.ylabel('F1 Score')
    plt.show()"""
```

```python
opt_depth,opt_est,tr_f1,cv_f1,depth,estimator= RF_tune(X,Y)
```

```python
# For train f1
data_plot = {'Depth':pd.Series(depth),'Estimator':pd.Series(estimator),'F1':pd.Series(tr_f1)}
df_plot = pd.DataFrame(data_plot)
df1_plot = df_plot.pivot("Depth", "Estimator", "F1")
ax = sns.heatmap(df1_plot,annot=True)
plt.show()
```

```python
# For validation f1
data_plot = {'Depth':pd.Series(depth),'Estimator':pd.Series(estimator),'F1':pd.Series(cv_f1)}
df_plot = pd.DataFrame(data_plot)
df1_plot = df_plot.pivot("Depth", "Estimator", "F1")
ax = sns.heatmap(df1_plot,annot=True)
plt.show()
```

```python
#getting the optimal
f1_RF = max(cv_f1)
print(opt_depth,opt_est,f1_RF)
```

```python
#retrain on full data
clf = RandomForestClassifier(max_depth = opt_depth, n_estimators = opt_est)
    
#Fit the model
clf.fit(X,Y)
    
#Compute accuracy on the training set
pred = clf.predict(test_data)
```

```python
d = {'employee_id': Emp_ID, 'is_promoted': pred}
upload_RF = pd.DataFrame(d)
upload_RF.to_csv("upload_RF.csv",index=False)
```
